```python
import pandas as pd
from IPython.display import display

# Specify the filepaths
filepath_no_damage = r'C:\Users\200408\repos\HeroysundBridge-ML\Utilities\Diana_scripts/diana_no_damage_df.parquet'

filepath_cable_2N = r'C:\Users\200408\repos\HeroysundBridge-ML\Utilities\Diana_scripts/diana_cable_2N_df.parquet'
filepath_cable_1NB = r'C:\Users\200408\repos\HeroysundBridge-ML\Utilities\Diana_scripts/diana_cable_1NB_df.parquet'

filepath_diana_data = r'C:\Users\200408\repos\HeroysundBridge-ML\Utilities\Diana_scripts/filtered_diana_data.parquet'

# Open the Parquet files
no_damage = pd.read_parquet(filepath_no_damage)

cable_2N = pd.read_parquet(filepath_cable_2N)
cable_1NB = pd.read_parquet(filepath_cable_1NB)

diana_data = pd.read_parquet(filepath_diana_data)

# Create a DataFrame with 'pt100_temperature_mean' and 'average_global_radiation'
diana_weather = diana_data[['pt100_temperature_mean', 'average_global_radiation']]

# Create a DataFrame with 'point_1_n_mean' to 'point_6_s_mean'
real_sensordata = diana_data[['Point_1_N_mean', 'Point_2_N_mean', 'Point_3_N_mean', 'Point_4_S_mean', 'Point_5_S_mean', 'Point_6_S_mean']]

# Display the DataFrames
display(no_damage)
display(cable_2N)
display(cable_1NB)
display(diana_weather)
display(real_sensordata)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_dataframe(df, title):
    # Create a new column 'hours' with a range from 0 to 199
    df['hours'] = np.arange(0, 200)

    # Set 'hours' as the index of the DataFrame
    df.set_index('hours', inplace=True)

    # Plot the DataFrame
    df.plot()

    # Set the x and y labels
    plt.xlabel('Time [Hours]', fontsize=16)
    plt.ylabel('Strain [Âµm/m]', fontsize=16)
    plt.yticks(fontsize=14)
    plt.xticks(fontsize=14)

    # Set the title
    plt.title(title, fontsize=16)

    # Place the legend outside of the plot
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=14)

    # Show the plot
    plt.show()

# Plot the DataFrames
plot_dataframe(no_damage, 'Diana strain simulation, surgate model whit no damage')
plot_dataframe(cable_2N, 'Diana simulation removed cable 2N')
plot_dataframe(cable_1NB, 'Diana simulation removed cable 1NB')
plot_dataframe(real_sensordata, 'Recorded sensor strain data')
```

```python
import matplotlib.pyplot as plt



def plot_columns(df1, df2):
    # Loop through the first six columns
    for i in range(6):
        # Get the column name
        column_name1 = df1.columns[i]
        column_name2 = df2.columns[i]

        # Create a new figure
        plt.figure()

        # Plot the column from the first DataFrame
        df1[column_name1].plot(label='Diana simulation, no damage')

        # Plot the column from the second DataFrame
        df2[column_name2].plot(label='Recorded sensor strain data')

        # Add a legend with fontsize 16
        plt.legend(fontsize=16)

        # Set the title of the plot to the column name with fontsize 16
        plt.title(f"{column_name1} ", fontsize=16*1.5)

        # Set the x and y labels
       
        plt.xlabel('Time [Hours]', fontsize=16*1.5)
        plt.ylabel('Strain [Âµm/m]', fontsize=16*1.5)

        plt.yticks(fontsize=14*1.5)
        plt.xticks(fontsize=14*1.5)

        # Show the plot
        plt.show()




def plot_columns3(df1, df2, df3):
    # Get the minimum number of columns among the three dataframes
    min_columns = min(len(df1.columns), len(df2.columns), len(df3.columns))

    # Loop through the first min_columns columns
    for i in range(min_columns):
        # Get the column name
        column_name1 = df1.columns[i]
        column_name2 = df2.columns[i]
        column_name3 = df3.columns[i]

        # Create a new figure
        plt.figure()

        # Plot the column from the first DataFrame
        df1[column_name1].plot(label='No damage')

        # Plot the column from the second DataFrame
        df2[column_name2].plot(label='Removed cable 2N')

        # Plot the column from the third DataFrame
        df3[column_name3].plot(label='Removed cable 1NB')

        # Add a legend with fontsize 16
        plt.legend(fontsize=16)

        # Set the title of the plot to the column name with fontsize 16
        plt.title(f"{column_name1} ", fontsize=16*1.5)

        # Set the x and y labels with fontsize 16
        plt.xlabel('Time [Hours]', fontsize=16*1.5)
        plt.ylabel('Strain [Âµm/m]', fontsize=16*1.5)

        plt.yticks(fontsize=14*1.2)
        plt.xticks(fontsize=14*1.2)

        # Place the legend outside of the plot
        plt.legend(bbox_to_anchor=(0.5, 0.3), loc='upper left', fontsize=14)

        # Show the plot
        plt.show()

def plot_columns3X(df1, df2, df3):
    # Get the minimum number of columns among the three dataframes
    min_columns = min(len(df1.columns), len(df2.columns), len(df3.columns))

    # Create a figure with min_columns subplots
    fig, axs = plt.subplots(min_columns, 1, figsize=(10, min_columns*3))

    # Loop through the first min_columns columns
    for i in range(min_columns):
        # Get the column name
        column_name1 = df1.columns[i]
        column_name2 = df2.columns[i]
        column_name3 = df3.columns[i]

        # Plot the column from the first DataFrame
        axs[i].plot(df1[column_name1], label='No damage')

        # Plot the column from the second DataFrame
        axs[i].plot(df2[column_name2], label='Removed cable 2N')

        # Plot the column from the third DataFrame
        axs[i].plot(df3[column_name3], label='Removed cable 1NB')

        # Add a legend with fontsize 16
        axs[i].legend(fontsize=16)

        # Set the title of the plot to the column name with fontsize 16
        axs[i].set_title(f"{column_name1} ", fontsize=16)

        # Set the x and y labels with fontsize 16
        axs[i].set_xlabel('Time [Hours]', fontsize=16)
        axs[i].set_ylabel('Strain [Âµm/m]', fontsize=16)

        axs[i].tick_params(axis='both', which='major', labelsize=14)

        # Place the legend outside of the plot
        axs[i].legend(bbox_to_anchor=(0.5, 0.6), loc='upper left', fontsize=14)

    # Adjust the layout so that there is no overlap between subplots
    plt.tight_layout()
    plt.show()

        
```

```python
import matplotlib.pyplot as plt

def plot_columnsX(df1, df2):
    # Define the number of rows for the subplot grid (6 plots, 1 column)
    num_plots = 6
    cols = 1  # Set to 1 for a single column of plots
    rows = num_plots  # Each plot gets its own row

    # Create a figure with subplots
    fig, axes = plt.subplots(rows, cols, figsize=(10, 15))  # Adjust the figure size
    axes = axes.flatten()  # Flatten the axes array for easy iteration

    # Loop through the first six columns
    for i in range(num_plots):
        # Get the column names
        column_name1 = df1.columns[i]
        column_name2 = df2.columns[i]

        # Plot the column from the first DataFrame
        axes[i].plot(df1.index, df1[column_name1], label='Diana simulation, no damage')

        # Plot the column from the second DataFrame
        axes[i].plot(df2.index, df2[column_name2], label='Recorded sensor strain data')

        # Set subplot title
        axes[i].set_title(f"{column_name1}", fontsize=16)  # Adjust title fontsize

        # Set the x and y labels with fontsize
        axes[i].set_xlabel('Time [Hours]', fontsize=16)
        axes[i].set_ylabel('Strain [Âµm/m]', fontsize=16)

        # Set tick parameters for x and y axes
        axes[i].tick_params(axis='x', labelsize=16)
        axes[i].tick_params(axis='y', labelsize=16)

        # Add a legend with fontsize
        axes[i].legend(fontsize=14)

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Show the plot
    plt.show()
```
